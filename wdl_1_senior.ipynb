```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
 def fun(lst): 
         links = list(lst.linkid) 
         if len(links) > 1: 
             for i in range(len(links)): 
                 G.add_edge(links[i], links[(i+1) % len(links)])
            
```

```python
G = nx.Graph()
```

```python
df_routes = pd.read_csv('BusRoutes.txt', encoding= 'unicode_escape',sep="|")
```

```python
G.add_nodes_from(df_routes.linkid.unique())
```

```python
df_routes.groupby('IDRoute').apply(fun)
```
